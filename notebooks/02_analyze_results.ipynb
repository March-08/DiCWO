# 02 — Analyze Results

Load and inspect individual experiment results.

```python
import sys
sys.path.insert(0, '..')

import json
from pathlib import Path
from src.runner.comparison import find_runs, load_run
```

```python
# List all runs
runs = find_runs('../results')
for r in runs:
    print(r.name)
```

```python
# Load a specific run
if runs:
    data = load_run(runs[-1])  # Most recent
    print('Metadata:', json.dumps(data.get('metadata', {}), indent=2))
    print('\nMetrics:', json.dumps(data.get('metrics', {}).get('totals', {}), indent=2))
```

```python
# View evaluation scores
if runs:
    eval_data = data.get('evaluation', {})
    if eval_data:
        print(json.dumps(eval_data, indent=2))
    else:
        print('No evaluation data found')
```

```python
# View artifacts (outputs)
if runs:
    artifacts = data.get('artifacts', {})
    for key, value in artifacts.items():
        print(f'\n{"="*60}')
        print(f'Artifact: {key}')
        print(f'{"="*60}')
        print(str(value)[:1000])
```
